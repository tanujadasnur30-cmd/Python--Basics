```python
i=0
while i<=10:
  print(i,end=" ")
  i=i+1
```

```python
start=int(input())
end=int(input())
if start<end:
  while start<=end:
    print(start,end="  ")
    start+=1
else:
  while start>=end:
    print(start,end="  ")
    start-=1

```
